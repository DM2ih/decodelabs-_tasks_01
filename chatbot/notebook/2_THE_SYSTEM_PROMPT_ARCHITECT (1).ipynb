#**MEMORY+** **PROMPTING**

```python
# =========================================================
# SIMPLE MEMORY LIST
# =========================================================

chat_memory = []
```

```python
def ask_aurora(query):

    messages = []

    # -----------------------------------
    # Add system prompt
    # -----------------------------------

    messages.append(
        SystemMessage(content=SYSTEM_PROMPT)
    )

    # -----------------------------------
    # Add few-shot examples (corrected conversion)
    # -----------------------------------

    for example in FEW_SHOT_EXAMPLES:

        role = example["role"]
        text = example["parts"][0]["text"]

        if role == "user":
            messages.append(HumanMessage(content=text))

        elif role == "model":
            messages.append(AIMessage(content=text))

    # -----------------------------------
    # Add previous memory
    # -----------------------------------

    messages.extend(chat_memory)

    # -----------------------------------
    # Add current query
    # -----------------------------------

    user_message = HumanMessage(content=query)

    messages.append(user_message)

    # -----------------------------------
    # Get response from Gemini
    # -----------------------------------

    response = model_llm.invoke(messages)

    # -----------------------------------
    # Store in memory
    # -----------------------------------

    chat_memory.append(user_message)

    chat_memory.append(
        AIMessage(content=response.content)
    )

    return response.content
```

#**VALIDATION**

```python
BANNED_WORDS = [
    "cheap",
    "budget airline",
    "competitor"
]

def validate_response(text):

    for word in BANNED_WORDS:

        if word.lower() in text.lower():
            return False

    return True
```

```python
import time

# Reset memory to start fresh
chat_memory = []

while True:
    # 1. Force a small delay to let Colab finish rendering the previous message
    time.sleep(0.5)

    print("=" * 70)
    query = input("You: ").strip()

    if query.lower() in ["q", "exit", "quit"]:
        print("\nExiting chat...")
        break

    if not query:
        continue

    try:
        ai_reply = ask_aurora(query)

        if validate_response(ai_reply):
            print("\nAurora:\n")
            print_md(ai_reply)
        else:
            print("\n[System]: Response blocked due to policy violation.")

    except Exception as e:
        print(f"\n[Critical Error]: {e}")
        print("Check if your model name is 'gemini-1.5-flash' and your API key is valid.")
        break
```
